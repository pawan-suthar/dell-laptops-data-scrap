## import Modules


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

## send req to flipkart


```python
website = 'https://www.flipkart.com/search?q=dell+laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1'

response = requests.get(website)

print(response.status_code)
```

## Object


```python
soup = BeautifulSoup(response.content, 'html.parser')
```

## result

```python
results = soup.find_all('div', {'class' : '_2kHMtA'})
len(results)
```

## target dell laptops data

```python
## laptop name
results[0].find('div', {'class':'_4rR01T'}).get_text()
```

```python
## laptop processor
results[0].find('li').get_text()
```

```python
# specs
results[0].find('div',{'class':'fMghEO'}).get_text()
```

```python
## laptop price
results[0].find('div',{'class':'_30jeq3 _1_WHN1'}).get_text()
```

```python
N_ame = []
P_rice = []
Specs = []

for i in range(1,6):
    website = 'https://www.flipkart.com/search?q=dell+laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+ str(i)
    # request to website
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    results = soup.find_all('div', {'class' : '_2kHMtA'})
    
    for result in results: 
        try:
            N_ame.append(result.find('div', {'class':'_4rR01T'}).get_text())
        except:
            N_ame.append('N/A')
        try: 
            P_rice.append(result.find('div',{'class':'_30jeq3 _1_WHN1'}).get_text())
        except:
            P_rice.append('N/A')
        try:
            Specs.append(results[0].find('div',{'class':'fMghEO'}).get_text())
        except:
            Specs.append('N/A')
```

## create data

```python
laptops = pd.DataFrame({'name':N_ame, 'price':P_rice, 'specs':Specs})
laptops
```

```python
laptops.to_excel('dell laptops.xlsx', index=False)
```
